```python
#Fazendo uma cópia do dataframe

df1 = df.copy()
```

```python
#Depois de fazer a cópia, vamos remover as colunas que não vamos precisar para análise:

# axis=1 -> apaga a coluna
# axis=0 -> apaga a linha

df1 = df1.drop(["Unnamed: 0", "designation", "region_1", "region_2", "taster_name", "taster_twitter_handle"],axis=1)
df1.head()
```

```python
#Ainda temos o problema dos dados ausentes para resolver

df1.info()
```

```python
#Tamanho da base de dados:

df1.shape

#(129971, 8)
#(129971 registro em 8 colunas)
```

```python
#Cálculo de proporção na coluna de preço (que é bastante importante para nossa análise) e apresenta dados ausentes:

print("{:.0%}".format(df1["price"].count() / df1.shape[0]))

#OBS: Apagando todos os registro em brancos ficamos com 93% da base de dados
```

```python
#Filtrando por país e uvas (country e variety) que tambem contém dados nulos

df1 = df1.dropna(subset=["country", "price", "variety"])
df1.info()
```

```python
#ANÁLISE EXPLORATÓRIA
#Sempre que houver dúvidas por onde começar, visualize os dados!
#Vamos olhar para a distribuição dos pontos dados durante as avaliações dos vinho, parece um bom ponto de partida.

plt.figure(figsize=(10, 7))
plt.hist(data=df1, x='points',
         bins=np.arange (80, df1['points'].max()+1, 1),
         color = 'purple')
plt.xlim(80,100)
plt.title('Distribuição dos Pontos', fontsize=16)
plt.xlabel('Pontos')
plt.ylabel('Frequencia')

#OBS: Existe uma distribuição quase normal dos pontos variando a escala entre 80 e 100.
```

```python
#Ver quais paises estão no Top 10 que mais produzem vinhos.

country_top_10 = df1["country"].value_counts()[:10]
country_top_10
```

```python
#Com base nesse filtro vamos gerar um gráfico para comparar os paises que mais tiveram vinhos avaliados:

plt.figure(figsize=(10, 7))
country_top_10.plot(kind='bar', color = 'purple')
plt.title('Número de Classificações de Vinhos por País', fontsize = 16)
plt.xlabel('País')
plt.ylabel('Total de Avaliações')
```

```python
#Com base nessa ordenação do gráfico acima, vamos investigar quantos paises produzem vinho de melhor qualidade de acordo coma as classificações.

melhor_qualidade = df1.groupby("country")['points'].agg("mean")
melhor_qualidade = melhor_qualidade.sort_values(ascending=False)[:10]

#Converter em uma Serie o DataFrame usando o método to_frame()

melhor_qualidade.to_frame().style.background_gradient(cmap='Purples', high=0.5)
```

```python
#Gerando um gráfico desse filtro:

y = melhor_qualidade.index.to_series()
x = melhor_qualidade #Acessa os valores em uma serie direta

plt.figure(figsize=(10, 7))
plt.xlim(80, 92)
sns.barplot(x=x, y=y, palette='Purples_r')
plt.title('Top 10 média de vinhos de melhor qualidade por país', fontsize = 16)
plt.xlabel('Média de pontos')
plt.ylabel('País')

#OBS: Ao verificar a média de pontos, podemos perceber que o país com melhor avaliação (que produziu o vinho de melhor qualidade) foi a Inglaterra.
#Isso ocorre, porque talvez haja menos vinhos na Inglaterra no nosso DataFrame e eles tenham uma pontuação alta, representa uma aumento na média.
```

```python
#Vale a pena investigarmos isso mais a fundo, podemos olhar para os paises com pontuação máxima acima de 95 (vinho excelentes).

# Número de vinhos por país com classificações mais altas de 95-100

high_scores = df1[df1["points"]>=95]
high_scores["country"].value_counts()
```

```python
# Os paises com classificação masi alta 95-100 pontos, diferem do nosso gráfico de pontuação mais alta com base nas médias.
#Podemos criar um gráfico tipo boxplot que permitirá ver o máximo, o mínimo e a média das pontuações por cada país.
#Como a base de dados apresenta mais vinhos avaliadosdos EUA, é provável que haja uma grande variedade de vinhos do país, enquanto a Inglaterra, que teve a melhor pontuação média, apresenta menos vinhos na base de dados, o que resulta em uma comparação injusta.

plt.figure(figsize=(20, 10))
sns.boxplot(data=df1, x='points', y='country', color = 'Purple')
plt.title('Pontos por País', fontsize = 16)
plt.xlabel('Pontos')
plt.ylabel('País')

#OBS: Embora os vinhos da Inglaterra tenham os pontos mais altos em média, os vinhos de Portugal, EUA, Itália, França e Austrália, receberam uma pontuação perfeita de 100.
# Há mais vinhos desses paises na base de dados, portanto eles receberam uma gama maior de pontos.
```

```python
#Investigação do preço de vinho nos paises:

average_price = df1.groupby("country")['price'].mean() #Calcula a média de cada país na columa preço
average_price = average_price.sort_values(ascending=False) #Valores em ordem descendente
y = average_price.index.to_series()
x = average_price.values

plt.figure(figsize=(12, 8))
sns.barplot(x=x, y=y, palette='Purples_r')
plt.title('Preço médio da garrafa de vinho por país', fontsize = 16)
plt.xlabel('Preço médio ($)')
plt.ylabel('País')

#OBS: A Suiça, produz o vinho mais caro, com o preço médio da garrafa de mais de $80.
# Já o vinho da Ucrânia, está no outro extremo, com preço médio mais baixo, $10 a garrafa.
```

```python
#Investigando se o preço está relacionado a qualidade do vinho.
#A utilização de histrograma para essa tarefa parece um bom plano:

#Distribuição normal
df1['price'].describe()
```

```python
#Distribuição logarítmica
np.log10(df1["price"].describe())
```

```python
#Agora utilizando os dados com a transformação logarítmica para o nosso histograma:

plt.figure(figsize=(12, 8))
bins = 10 ** np.arange(0.5, 3.5 + 0.05, 0.05)
ticks = [1, 3, 10, 30, 100, 300, 1000, 3000, 10000]
plt.hist(data=df1, x='price', bins=bins, color='Purple')
plt.title('Distribuição dos preços dos vinhos', fontsize = 16)
plt.xscale('log')
plt.xticks(ticks, ticks)
plt.xlabel('Preços($)')
plt.ylabel('Frequencia')

#OBS: Agora é possível ver uma distribuição (quase normal).
# Existe um salto na região dos 10 dólares e um pico mais visível em pouco mais de 20 dólares.
# Depois temos outro pico acima de $50.
```

```python
#Agora podemos plotar um gráfico do tipo scatter para visualizar se existe um correlação entre a pontuação e o valor da garrafa.

x = [1, 3, 10, 30, 100, 300, 1000, 3000, 10000]
plt.figure(figsize=(12, 8))
sns.regplot(x='price', y='points', data=df1, color='Purple', logx=True, scatter_kws={'alpha':0.3})
plt.title('Pontos vs Preço', fontsize = 14)
plt.xscale('log')
plt.xticks(x, x)
plt.xlabel('Preço ($)')
plt.ylabel('Pontos')

#OBS: O gráfico nos mostra uma relação positiva entre as duas variáveis (pontuação e preço), ou seja, vinhos mais caros tendem a receber melhores avaliações.
# Também é possível observar dois pontos mais afastados, um na região dos $4 e outro pouco depois de $3000.
```

```python
#Listar esses dois valores para descobrir quais vinhos são esses, buscando pelo valor máximo e mínimo:

#Valor mínimo:

df1[df1["price"]==df1["price"].min()][:1]
```

```python
#Valor máximo:

df1[df1["price"]==df1["price"].max()][:1]

#OBS: Temos um vinho Espanhol com 85 pontos de classificação que custa $4 a garrafa e um vinho Francês de 88 pontos que custa $3.300.
```

```python
#Filtrar uma amostra de 20 melhores vinhos e ver como eles pontuam.

df1["variety"].value_counts()[:20]
```

```python
#Gráfico com boxplot:

plt.figure(figsize=(20, 18))
sns.boxplot(data=df1,
            x='variety',
            y='points',
            color = 'Purple',
            order=df1['variety'].value_counts().iloc[:20].index)
plt.title('Distribuição de pontos por variedade', fontsize = 16)
plt.xlabel('Variedade (uvas)')
plt.ylabel('Pontos')
plt.xticks(rotation=90)

#Listando o top 20 melhores vinhos (com melhor classificação) na base de dados, vemos que a variedade Pinot Noir tende a ter o melhor desempenho, com a pontuação média mais alta.
```

```python
#Já sabemos que a qualidade está fortemente relacionada ao preço, mas nem todos os clientes podem gastar 3000 dólares em uma garrafa de vinho toda semana.
#Nossa tarefa agora é descobrir o melhor custo-beneficio em algumas faixas de preço.

# Até $15
# Entre $15 e $30
# Entre $30 e $50
# Acima de $50

# top 10 custando até $15
under_15 = df1.drop(["description", "province", "winery"], axis=1)
under_15 = under_15[under_15["price"] <= 15]
under_15 = under_15.sort_values(by="points", ascending=False)[:10]
under_15.style.background_gradient(cmap='YlOrRd', high=0.5, subset=['price'])

#OBS: Com base nos dados, uma boa escolha aqui seria um vinho Português de 94 pontos por $13, Quinta dos Murças 2011 Assobio Red (Douro)
```

```python

# top 10 com valor entre $15 e $30
between_15_and_30 = df1.drop(["description", "province", "winery"], axis=1)
between_15_and_30 = between_15_and_30.query("price > 15 and price <= 30")
between_15_and_30 = between_15_and_30.sort_values(by="points", ascending=False)[:10]
between_15_and_30.style.background_gradient(cmap='YlOrRd', high=0.5, subset=['price'])
```

```python
# top 10 com valor entre $30 e $50
between_15_and_30 = df1.drop(["description", "province", "winery"], axis=1)
between_15_and_30 = between_15_and_30.query("price > 30 and price <= 50")
between_15_and_30 = between_15_and_30.sort_values(by="points", ascending=False)[:10]
between_15_and_30.style.background_gradient(cmap='YlOrRd', high=0.5, subset=['price'])
```

```python
# top 10 com valor acima de $50
above_50 = df1.drop(["description", "province", "winery"], axis=1)
above_50 = above_50.query("price > 50")
above_50 = above_50.sort_values(by="points", ascending=False)[:10]
above_50.style.background_gradient(cmap='YlOrRd', high=0.5, subset=['price'])

#OBS: Por $80 é possivel experimentar o sabor de um vinho de 100 pontos, com duas opções: Charles Smith 2006 Royal City Syrah e Cayuse 2008 Bionic Frog Syrah, ambos dos EUA.
```

```python
#Descobrir quais palavras são mais utilizadas para descrever um vinho.
#Para isso podemos utilizar a biblioteca wordcloud que separa palvra em um texto e nos filtrar as de maior ocorrência, gerando uma nuvem de palavras.

#Primeiro passo importar a biblioteca e combinar todas as descrições das avaliaçôes em um só texto:
from PIL import Image
from wordcloud import WordCloud, STOPWORDS

text = df1['description'][1]
text = ' '.join(i for i in df1['description'])
len(text)
```

```python
#Temos agora um imenso texto agrupado.
#A biblioteca disponibiliza uma funcionalidade chamada stopwords que são uma lista de palavras que serão eliminadas no processo de criação de nossa nuvem de palavras.

stopwords = set(STOPWORDS)
stopwords.update(["drink", "now", "wine", "flavor", "flavors"])
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="white",
                      width=2048,
                      height=1024,
                      max_words=1000).generate(text)
```

```python
#Com a nuvem de palavras prontas, agora só precisamos plotar um imagem exibindo as palavras mais utilizadas nas descrições de vinho:

plt.figure(figsize=(20, 18))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
