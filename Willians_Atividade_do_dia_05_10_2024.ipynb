Entendendo o Dataset:

Rank: Se trata da posição daquele determinado jogo no ranking de vendas;

Name: Se trata do título do jogo;

Platform: Se trata do console em que o jogo foi lançado;

Year: Se refere ao ano de lançamento daquele título;

Genre: Se trata do gênero daquele jogo;

Publisher: Empresa responsável pela publicação do jogo;

NA_Sales: Vendas na América do Norte, em milhões de dólares;

EU_Sales: Vendas na Europa, em milhões de dólares;

JP_Sales: Vendas no Japão, em milhões de dólares;

Other_Sales: Vendas no em outras regiões além de Japão, Europa e América do Norte;

Global_Sales: Total de vendas mundiais.

```python
# 1. Importe o dataset para o Google Colab:
import pandas as pd

# Subindo o arquivo colab:
from google.colab import files
uploaded = files.upload()
```

```python
# Ler o dataset:
df = pd.read_csv('vgsales.csv')
```

```python
# Exibir primeiras linhas do dataframe:
df.head()
```

```python
# 2. Exploração dos dados
# Verificando presença de valores nulos:
df.isnull().sum()
```

```python
# 3. Tratamento dos dados faltantes:
df['Year'].fillna('Desconhecido', inplace=True)
df['Publisher'].fillna('Desconhecido', inplace=True)
```

```python
# 4. Conversão de tipos de dados:
df['Year'] = df['Year'].astype('category')
```

```python
# 5. Exploração Estatística dos Dados:
# Estatística descritivas para variáveis numéricas:
df.describe()
```

```python
# Frequência de valores para variáveis categóricas:
df['Platform'].value_counts()
```

```python
# 3. Criação de Gráficos (Visualização dos Dados)
# Gráficos para variáveis numéricas:

import matplotlib.pyplot as plt
import seaborn as sns

# 1)Histograma - Visualizar a distribuição de uma variável numérica:
# Visualizando histograma com frequência de gênero mais jogados:

sns.histplot(df['Genre'], bins=10, kde=True)
plt.ylabel('Frequência', size = 13)
plt.xlabel('Gênero', size = 13)
plt.title('Frequência de gênero mais jogado', size = 15)
plt.yticks(fontsize=12)
plt.xticks(fontsize=12, rotation=80)
plt.show()

```

```python
# 2)Boxplot - Detectar outliers e a dispersão dos dados:
sns.boxplot(x=df['Rank'])
plt.title('Ranking de vendas')
plt.show()
```

```python
# Gráficos para variáveis categóricas:
# 1) Gráfico de Barras - Visualizar a contagem de categorias:

sns.countplot(x='Platform', data=df)
plt.ylabel('Jogos Lançados', size = 13)
plt.xlabel('Consoles', size = 13)
plt.title('Consoles que mais jogos lançaram', size = 15)
plt.yticks(fontsize=12)
plt.xticks(fontsize=12, rotation=80)
plt.show()
```

```python
# 2)Gráfico de Pizza - Proporções de categorias:
df['Genre'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Proporção de Gêneros')
plt.ylabel('')
plt.show()
```

```python
# Gráfico relacionados:
# 1) Gráfico de Dispersão - Verificar correlação entre duas variáveis numéricas:

sns.scatterplot(x='NA_Sales', y='EU_Sales', data=df)
plt.title('Correlação entre as vendas da América do Norte x Europa')
plt.xlabel('Vendas na América do Norte')
plt.ylabel('Vendas na Europa')
plt.show()
```
