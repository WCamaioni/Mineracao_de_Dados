# 01 - Criação e Manipulação de Arrays

Crie um array Numpy com 10 números inteiros aleatórios entre 1 e 20 e extraia os elementos pares.

```python
import numpy as np

array = np.random.randint(1, 20, size=10)

elementos_pares = array[array % 2 == 0]

print("Array original:", array)
print("Array com elementos pares:", elementos_pares)
```

# 02 - Operações Aritméticas com Arrays

Crie dois arrays de 10 números inteiros aleatórios entre 0 e 50 e realize operações de soma, subtração, multiplicação e divisão elemento a elemento.

```python
array1 = np.random.randint(0, 51, size=10)
array2 = np.random.randint(0, 51, size=10)

# Operações de soma, subtração e multiplicação
soma = array1 + array2
subtracao = array1 - array2
multiplicacao = array1 * array2

# Para evitar divisão por zero, substituímos valores zero no array2 por 1
array2_nonzero = np.where(array2 == 0, 1, array2)
divisao = array1 / array2_nonzero

print("Array 1:", array1)
print("Array 2:", array2)
print("")
print("Soma:", soma)
print("Subtração:", subtracao)
print("Multiplicação:", multiplicacao)
print("Divisão:", divisao)
```

# 08 - Filtragem de Dados

Filtre o DataFrame criado no exercício anterior para mostrar apenas as vendas onde a quantidade é maior que 10.

```python
import numpy as np
import pandas as pd

# Produtos
produtos = np.array(['Arroz', 'Feijão', 'Café', 'Farinha', 'Bolacha','Margarina'])

# Quantidade entre 1 e 20
quantidade = np.random.randint(0, 21, size=6)

# Preço entre 10 e 100
preco = np.random.randint(10, 101, size=6)

# Vendas entre 0 e 21
venda = np.random.randint(0, quantidade + 1)

# Criando o DataFrame
df = pd.DataFrame({
    'Produtos': produtos,
    'Quantidade': quantidade,
    'Preço(R$)': preco,
    'Vendas': venda 
})

# Filtrando o DataFrame para mostrar apenas as vendas onde a quantidade é maior que 10
df_filtered = df[df['Vendas'] > 10] 

print("DataFrame Completo:")
print(df)

print("\nDataFrame Filtrado (Quantidade > 10):")
print(df_filtered)

# Verificar se há algum item com vendas maiores que 10
if df_filtered.empty:
    print("\n***Não houve vendas maiores que 10 unidades!!!***")
```

# 11 - Gráfico de Linhas

Crie um gráfico de linha que mostra a variação de preços de um produto fictício ao longo de 12 meses.

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados fictícios
meses = np.array(['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'])
precos = np.array([50, 52, 51, 53, 55, 54, 56, 57, 58, 60, 59, 61])

# Criar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(meses, precos, marker='o', linestyle='-', color='b', label='Preço')

# Adicionar título e rótulos aos eixos
plt.title('Variação de Preços do Produto ao Longo de 12 Meses')
plt.xlabel('Mês')
plt.ylabel('Preço Médio de Livros (R$)')

# Adicionar uma grade para melhor visualização
plt.grid(True)

# Mostrar a legenda
plt.legend()

# Rotacionar os rótulos dos eixos X para melhor leitura
plt.xticks(rotation=45)

# Mostrar o gráfico
plt.tight_layout()
plt.show()

```

# 17 - Gráfico de Dispersão com Seaborn

Recrie o gráfico de dispersão da relação entre altura e peso usando Seaborn, incluindo uma linha de regressão.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Gerar dados fictícios para altura e peso
np.random.seed(0)  # Para garantir a reprodutibilidade
altura = np.random.normal(loc=170, scale=10, size=100)  # altura média de 170 cm, desvio padrão de 10 cm
peso = np.random.normal(loc=70, scale=15, size=100)  # peso médio de 70 kg, desvio padrão de 15 kg

# Criar um DataFrame com os dados
df = pd.DataFrame({
    'Altura': altura,
    'Peso': peso
})

# Criar o gráfico de dispersão com linha de regressão
sns.lmplot(x='Altura', y='Peso', data=df, aspect=1.5)

# Adicionar título e labels
plt.title('Gráfico de Dispersão da Relação entre Altura e Peso com Linha de Regressão')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')

# Mostrar o gráfico
plt.show()
```

# 19 - Heatmap

Crie um heatmap para mostrar a correlação entre diferentes variáveis em um DataFrame fictício.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crie um DataFrame fictício com variáveis aleatórias
np.random.seed(0)
data = np.random.rand(10, 5)  # 10 amostras, 5 variáveis
columns = ['A', 'B', 'C', 'D', 'E']
df = pd.DataFrame(data, columns=columns)

# Calcule a matriz de correlação
correlation_matrix = df.corr()

# Crie o heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap de Correlação')
plt.show()

```

# 20 - Pairplot

Crie um pairplot para explorar relações entre várias variáveis em um conjunto de dados fictício.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Definir o número de amostras
n_samples = 100

# Gerar dados fictícios
np.random.seed(0)  # Para reprodutibilidade

# Variáveis independentes
x1 = np.random.normal(loc=0, scale=1, size=n_samples)
x2 = np.random.normal(loc=5, scale=2, size=n_samples)
x3 = np.random.normal(loc=-3, scale=0.5, size=n_samples)

# Criar um DataFrame
df = pd.DataFrame({
    'Variable 1': x1,
    'Variable 2': x2,
    'Variable 3': x3,
    'Variable 4': x1 * x2 + np.random.normal(size=n_samples)  # Variável dependente com alguma relação
})

# Criar um pairplot
sns.pairplot(df)
plt.show()

```
